```python
data = {
    "salary": [9500, 10200, 9800, 10000, 9700, 10150, 9900, 10300, 9600, 10050]
} 
```

```python
df=pd.DataFrame(data)
```

```python
df.shape
```

```python
#Line Plot
plt.plot(df["salary"],color="red",marker="p",linestyle=":",linewidth="2")
plt.grid()
plt.show()
```

```python
#Histogram
plt.hist(df["salary"],bins=5,color="green")

plt.show()
```

```python
plt.boxplot(df["salary"])
plt.show()
```

```python
df.loc[9]=8500
plt.boxplot(df["salary"])
plt.show()
```

```python
df.drop(index=9,inplace=True)
```

Bivariate

```python
df.head()
```

```python
df["Age"]=[18, 20, 22, 21, 19, 23, 24, 20, 22]
df.head()
```

```python
#scatter plot
plt.scatter(df["Age"],df["salary"], color="orange")
plt.show()
```

```python
sort_age=df.sort_values("Age")
```

```python
#LINEPLOT
plt.plot(sort_age["Age"],df["salary"],color="red",marker="o",linestyle="-",linewidth="2")
plt.grid()
plt.show()
```

```python
#barchart
plt.bar(sort_age['Age'],df['salary'],color='green')
plt.show()
```

```python
#bivariate:numerical-categorical
hr_sal=df[df['dept']=="HR"]["salary"]
hr_sal
```

```python
it_sal=df[df['dept']=='IT']["salary"]
it_sal
```

```python
sales_sal=df[df['dept']=='Sales']["salary"]
sales_sal
```

```python
#boxplot
plt.boxplot([hr_sal,it_sal,sales_sal],labels=["hr","it","sales"])
plt.show()
```

```python
#piechart
salary_by_dept=df.groupby("dept")["salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept,labels=salary_by_dept.index,autopct="%1.2f",shadow=True,explode=[0,0.1,0])
plt.axis("equal")
plt.show()
```

```python
#barplot
hr_mean=sum(hr_sal)/len(hr_sal)
it_mean=sum(it_sal)/len(it_sal)
sales_mean=sum(hr_sal)/len(sales_sal)
```

```python
plt.bar(["hr","it","sales"],["hr_mean","it_mean","sales_mean"],color=["green","black","red"])
plt.grid()
plt.show()
```

```python
#Multivariate Analysis:Numerical
df.head() 
```

```python
df["experience"]=[0, 1, 2, 3, 4, 5, 6, 2, 3]
```

```python
df.head()
```

```python
#Bubble plot
plt.scatter(df["Age"],df["salary"],s=df["experience"]*20,color="pink")
plt.title("Age vs Salary vs Experience ")
plt.xlabel("age")
plt.ylabel("salary")
plt.grid()
plt.show()
```

```python
#2 numerical and 1 categorical: Scatter plot 
plt.scatter(df["Age"],df["salary"],c=df["dept"].map({"HR":"yellow","IT":"red","Sales":"green"}),label )
plt.xlabel("Age")
plt.ylabel("salary")
plt.legend()
plt.show()
```

```python
color={"HR":"yellow","IT":"red","Sales":"green"}
for dept, color in color.items():
    df_dept=df[df["dept"]==dept]
    plt.scatter(df_dept["Age"],df_dept["salary"],c=color,label=dept)
plt.legend()
```

Object oriented api

```python
#Lineplot
fig, axs=plt.subplots(1,3,figsize=(15,5))
sort_age
axs[0].plot(sort_age["Age"],df["salary"],color="red", marker="o",linewidth="2",markersize="2")
axs[0].grid()
axs[0].set_title("Line PLot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("salar")

#histogram
axs[1].hist(df["salary"],bins=3,color="red")
axs[1].set_title("histogram")
axs[1].set_xlabel("salary")
axs[1].set_ylabel("frequency")


#boxplot 
axs[2].boxplot(df["salary"])
axs[2].set_title("boxplot")
axs[2].set_xlabel("salary")

plt.show()
```

```python
data2 = {
    "Year": [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Sales": [50000, 55000, 60000, 58000, 62000, 65000, 70000, 75000, 80000],
    "Profit": [5000, 6000, 7500, 7000, 8000, 9000, 10000, 11000, 12000],
    "Experience": [0, 1, 2, 3, 4, 5, 6, 7, 8]
}
df2=pd.DataFrame(data2)
```

```python
plt.plot(df2["Year"],df2["Sales"],label="sales")
plt.plot(df2["Year"],df2["Profit"],label="profit")
plt.plot(df2["Year"],df2["Experience"],label="exp")
plt.title("financial analysis")
plt.xlabel("Year")
plt.ylabel("amount")
plt.legend()
plt.show()
```
