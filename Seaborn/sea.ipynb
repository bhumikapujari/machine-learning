```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g",hue="sex")
```

```python
sns.set_style("darkgrid")
sns.set_context("paper")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g",hue="island",palette="Dark2")
sns.despine(left=True, bottom=True)
```

```python
sns.scatterplot(data=penguins, x="species", y="body_mass_g",hue="island",palette="Dark2",style="sex",alpha =0.5)
```

```python
sns.stripplot(data=penguins, x="species", y="body_mass_g",hue="island",dodge=True,jitter=True)
```

```python
#swarm plot
sns.set_context("talk")
sns.swarmplot(data=penguins, x="species", y="body_mass_g",hue="island")
sns.despine()
```

```python
#histogram
sns.set_context("notebook")
sns.histplot(data=penguins, x="body_mass_g",hue="sex",multiple="stack")
```

```python
#regression plot
sns.regplot(data=penguins, x="flipper_length_mm", y="body_mass_g")
```

```python
#LinePlot
sns.lineplot(data=penguins, x="flipper_length_mm", y="body_mass_g",hue="island",style="species")
```

```python
#joinplot
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g")
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="scatter",hue="sex")
```

```python
#barplot
sns.barplot(data=penguins, x="species", y="body_mass_g",hue="sex",palette=["skyblue","pink"],estimator=np.median)
```

```python
#countplot
sns.countplot(data=penguins, x="species",hue="island",palette="Set2")
```

```python
#boxplot  
sns.boxplot(data=penguins, x="species", y="body_mass_g",hue="sex",palette="Set3")

```

```python
#violinplot
sns.violinplot(data=penguins, x="species", y="body_mass_g",hue="sex",split=True,palette="Set1")
#width of the violin :desity od the data at different values
#middle line : median
#Thick part : interquartile range (if inner="box")
#interquartile means the thirs quartile - first quartile
```

```python
sns.violinplot(data=penguins, x="species", y="body_mass_g")
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue ="island", size=3)
```

```python
#kde plot(kernal density estimation)
#in simple terms it is a smooth curve that shows the distribution of data
#It helps you see where the data is concentrated(peaks) and how it spreads
#smooth version of histogram
sns.kdeplot(data=penguins, x="body_mass_g", hue="species", fill=True)
```

```python
#Heatmap is a 2D color-codedd matrix that shows relationships or patterns between two variables
#commanly used to : show correlation between features
#-visualize confusion matrix (will be seen in ml)
#-highlight missing data

penguins.head()
```

```python
columns=["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
penguins[columns].corr()

```

```python
sns.heatmap(data=penguins[columns].corr(),annot=True,cmap="Blues",vmin=-0.2)
```

```python
#Rugplot
sns.rugplot(data=penguins, y="body_mass_g", hue="species",palette="Set2",height=0.3)
```

```python
#pairplot
sns.pairplot(data=penguins, hue="sex",palette="Set2",diag_kind="hist")
```

```python
#pairgrid: subplot grid for plotting pairwise relationships in a dataset
graph = sns.PairGrid(data=penguins, hue="sex",palette="Set2") 
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
graph.add_legend()
```
