```python
corr = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
           'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement',
           'yr_built', 'yr_renovated']].corr()

plt.figure(figsize=(8, 4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

The highest correlation with the price is sqft_living (bigger living area → higher price)
sqft_above, bathrooms and view also have good correlations (house size + bathrooms + view)


sqft_lot = 0.05 (almost no relation)
condition = 0.04 (very weak)
yr_built = 0.02 (almost no relation)

Strong correlations between features:-
sqft_living and sqft_above = 0.88 (VERY high)
sqft_living and bathrooms = 0.76
sqft_above and bathrooms = 0.69

Renovation Analysis

```python
df["Renovated"] = df["yr_renovated"] > 0

print(df["Renovated"].value_counts())
print(df.groupby("Renovated")["price"].mean())
print(df.groupby("Renovated")["price"].median())
```

```python
#Comparing sqft_living for Renovated vs Not Renovated
print(df.groupby("Renovated")["sqft_living"].mean())
print(df.groupby("Renovated")["sqft_living"].median())
```

```python
#barplot sqft_living
sns.barplot(x="Renovated", y="sqft_living", data=df, estimator="mean")
plt.title("Average Sqft Living: Renovated vs Not Renovated")
plt.show()
```

```python
#Boxplot of Price by Renovated (Outliers + Spread)
sns.boxplot(x="Renovated", y="price", data=df)
plt.title("Price Distribution: Renovated vs Not Renovated")
plt.show()
```

```python
#boxplot for yr_built for Renovated vs Not Renovated
sns.boxplot(x="Renovated", y="yr_built", data=df)
plt.title("Year Built: Renovated vs Not Renovated")
plt.show()
```

what to understand:-
-Look at the median line:
False (not renovated) median ≈ around 1985
True (renovated) median ≈ around 1958–1960
So renovated houses are typically built much earlier.

-Renovation happens mostly in old houses

-Renovated houses have a wider spread

The boxplot shows that renovated houses are generally older than non-renovated houses. The median year built for renovated houses is lower, indicating renovation is more common in older properties.
