```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [9500, 9700, 9900, 10100, 10300]

plt.plot(x, y, 
         color="green",
         linestyle="--",
         linewidth=2,
         marker="o",
         label="Salary")

plt.xlabel("Year")
plt.ylabel("Salary")
plt.title("Salary Growth")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

salary = [9500, 10200, 9800, 10000, 9700, 10150, 9900, 10300, 9600, 10050]

plt.hist(salary, 
         bins=6, 
         color="green", 
         edgecolor="black", 
         alpha=0.7, 
         label="Salary")

plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Salary Histogram")
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

salary = [9500, 10200, 9800, 10000, 9700, 10150, 9900, 10300, 9600, 10050]

plt.boxplot(salary,
            vert=True,
            labels=["Salary"],
            showfliers=True,
            patch_artist=True,
            boxprops=dict(color="green"),
            medianprops=dict(color="red"),
            whiskerprops=dict(color="black"),
            capprops=dict(color="black"),
            flierprops=dict(marker="o", color="blue"))

plt.ylabel("Salary")
plt.title("Salary Distribution")
plt.show()
```
